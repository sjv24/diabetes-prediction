# Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Collecting the data into a DataFrame

```python
df=pd.read_csv('/content/diabetes.csv')
```

From this we can analyse that, diabetic patients have an average insulin of 100 and a higher glucose, averaging upto to 141.25. People with a higher age are also at risk for diabetes.

In this case our target variable is **Outcome** (dependent variable) whereas other features are (independent variable)

```python
#separating target variable from dataset
X=df.drop(columns='Outcome', axis=1)
Y=df.Outcome
```

**Data Standardisation**

```python
scaler=StandardScaler()
std_df= scaler.fit_transform(X)
# reassigning X with standardized values
X= std_df
X
```

```python
print(X)
print()
print(Y)
```

# Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=100)
(X.shape, X_test.shape, X_train.shape)
# 154 records for test
# 614 records for train
```

Training the model

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

# Model Evaluation

Accuracy Score

```python
#training
x_train_prediction=classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction, Y_train )

print ("ACCURACY SCORE", training_data_accuracy)
```

```python
#testing
x_test_prediction=classifier.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction, Y_test )

print ("ACCURACY SCORE", test_data_accuracy)
```

A score above 75 is a good score

## Building a User-Interactive System



```python
def bmi_calc():
  height=float(input("Enter your height (metres): "))
  weight=float(input("Enter your weight (kgs): "))
  return weight/height**2

ch=input("Do you know your BMI? (y/n):  ")
if ch=='n' or ch=='N':
  bmi=bmi_calc()
  print("YOUR BMI IS:",bmi)


input_data=[]
input_questions=['How many times have you been pregnant? ',
                 'Enter glucose measurement: ',
                 'Enter blood pressure: ',
                 'Enter Skin Thickness: ',
                 'Enter Insulin: ',
                 "What is your BMI? ",
                 'Diabetes pedigree function: ',
                 'How old are you? ']

for i in range(len(input_questions)):
  x=float(input(input_questions[i]))
  input_data.append(x)

input_data_as_np_array= np.asarray(input_data)
input_data_reshaped= input_data_as_np_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshaped)

print()
prediction= classifier.predict(std_data)
if prediction == 1:
  print("This patient is predicted to have diabetes")
else:
  print("This patient is not predicted to have diabetes")
```
